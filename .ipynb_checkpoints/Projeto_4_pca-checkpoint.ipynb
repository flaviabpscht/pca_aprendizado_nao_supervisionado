## <a> Redução de Dimensionalidade com PCA! </a>

Reduzir a dimensionalidade utilizando análise de componentes principais (PCA).

```python
### <a>Primeiro, vamos analisar quantos componentes vamos usar (o menor número com maior variância "explicada")</a>
```

```python
from sklearn.decomposition import PCA
import matplotlib.cm as cm
```

```python
#remover as colunas demográficas para deixar somente o questionário (retirar também a variável mvliking)
colunas_demograficas = ['age', 'income', 'miles', 'numkids', 'female', 'educ', 'recycle', 'mvliking']

#testando o número de componentes e calculando a variância explicável de cada um

for numero_componentes in range(1, 9):
    pca = PCA(n_components = numero_componentes, random_state =42).fit(df_microvan.drop(colunas_demograficas, axis=1))
    print(f'{numero_componentes} componentes com variância explicada total de {sum(pca.explained_variance_ratio_)}: {pca.explained_variance_ratio_}')
  

```

```python
#opção por 5 componentes
pca = PCA(n_components = 5, random_state = 42).fit(df_microvan.drop(colunas_demograficas, axis = 1))
```

```python
#os valores de cada componente
pca.singular_values_
```

```python
df_microvan.shape
```

## <a>Análise dos Componentes</a>
Cada componente é criado como "parte" das variáveis originais. Ou seja, as variáveis originais compõe os componentes principais. Para saber o impacto de cada variável na projeção final, existe um atributo no objeto PCA que mostra esses valores.

## <a>Analysis of Components</a>
Each component is created as a "part" of the original variables. In other words, the original variables make up the principal components. To understand the impact of each variable on the final projection, there is an attribute in the PCA object that shows these values."

```python
pca.components_
```

```python
#mostrar a composição de cada componente e peso das variáveis originais
df_composicao_componentes = pd.DataFrame(columns=df_microvan.drop(colunas_demograficas, axis=1).columns)

for i in range(5):
    df_composicao_componentes.loc[i] = pca.components_[i]

df_composicao_componentes
```

```python
df_composicao_componentes.describe()
```

```python
df_composicao_componentes.info()
```

```python
#avaliação da maior influência positiva no primeiro argumento
df_composicao_componentes.loc[0].max()
```

```python
df_composicao_componentes.loc[0].argmax()
```

```python
#para avalaiar a composição de cada componente

for i in range(5):
    print(f'Componente {i+1} - Impactos positivos mais relevantes: ')
    for col in df_composicao_componentes.columns:
        valor = df_composicao_componentes.loc[i, col]
        if valor in df_composicao_componentes.loc[i].nlargest(6).values:
            print(f'      {dicionario_dados[col]}: {valor}')

    print(f'Componente {i+1} - Impactos negativos mais relevantes: ')
    for col in df_composicao_componentes.columns:
        valor = df_composicao_componentes.loc[i, col]
        if valor in df_composicao_componentes.loc[i].nsmallest(6).values:
            print(f'      {dicionario_dados[col]}: {valor}')

    print('\n')

```

#### <a>Analisando graficamente os pesos de cada variável nos componentes principais</a>

```python
plt.figure(figsize=(16, 8))
sns.heatmap(data=df_composicao_componentes, yticklabels=df_composicao_componentes.index+1)
```

Cada linha do dataframe df_composicao_componentes representa um componente principal e cada valor nas colunas representa o peso (ou coeficiente) atribuído a uma característica específica naquele componente principal. Esses pesos indicam a contribuição relativa de cada característica para aquele componente.

Embora todas as características originais estejam presentes no dataframe df_composicao_componentes, elas estão representadas em termos de como contribuem para os cinco primeiros componentes principais, que são as novas dimensões do espaço de características após a redução de dimensionalidade realizada pelo PCA.



Each row of the dataframe df_composicao_componentes represents a principal component, and each value in the columns represents the weight (or coefficient) assigned to a specific feature in that principal component. These weights indicate the relative contribution of each feature to that component.

Although all original features are present in the dataframe df_composicao_componentes, they are represented in terms of how they contribute to the first five principal components, which are the new dimensions of the feature space after the dimensionality reduction performed by PCA.
